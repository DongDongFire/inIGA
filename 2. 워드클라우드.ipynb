```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
#http://amueller.github.io/word_cloud/generated/wordcloud.WordCloud.html
#https://busy.org/@anpigon/2
```

```python
df=pd.read_excel('C:/Users/IGAW/Desktop/abc.xlsx',sheet_name='c')
keywords=df.set_index('Category').to_dict('dict')
keywords['Frequency']
```

```python
from wordcloud import WordCloud
import numpy as np
from PIL import Image
font_path='c:/windows/fonts/NanumGothic.ttf'
mask = np.array(Image.open('C:/Users/IGAW/Desktop/cloud.png'))
wordcloud = WordCloud(
    font_path = font_path,
    width = 800,
    height = 800,
    max_font_size=120,
    min_font_size=30,
    background_color='white',
    mask=mask,
    colormap="copper_r"
)
wordcloud = wordcloud.generate_from_frequencies(keywords['Frequency'])

fig = plt.figure(figsize=(9,9))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.show()
```
