```python
from selenium import webdriver as wd
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import requests
import re
import time
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import os
from datetime import datetime

options = Options()


# chrome_options.add_argument("--headless")
options.add_argument("--window-size=1920x1080")
options.add_argument("disable-gpu")
driver=wd.Chrome(executable_path = 'C:/Users/IGAW/chromedriver.exe',options=options)

url='https://search.naver.com/search.naver?sm=top_hty&fbm=1&ie=utf8&query=전국골프장'

driver.get(url)
driver.implicitly_wait(10)
res=driver.page_source

i=1
name=[]
category=[]
address=[]
phone=[]
fee=[]
while len(name)<=421:
    res=driver.page_source
    html=BeautifulSoup(res,'html.parser')
    
    for i in html.findAll('div',{'class':'cc_name'}):
        print(i.text)
        name.append(i.text)

    
    for k in html.findAll('div',{'class':'cc_member'}):
        category.append(k.text)
    
    address_list=html.findAll('div',{'class':'cc_address'})
    for add in address_list:
        address.append(add.text)

    for a in html.findAll('div',{'class':'cc_phone'}):
        phone.append(a.text)
    

    for b in html.findAll('div',{'class':'cc_greenfee'}):
        fee.append(b.text)
    
    if len(name)==8:
        driver.find_element_by_xpath("//*[@id='_golf_nav']/a").click()
    else:
        driver.find_element_by_xpath("//*[@id='_golf_nav']/a[2]").click()
    
    driver.implicitly_wait(10)
    
```

```python
df=pd.DataFrame({'명칭':name,'분류':category,'주소':address,'전화번호':phone,'그린피':fee
                })
```

```python
df.to_excel('전국골프장 리스트(네이버).xlsx')
```
